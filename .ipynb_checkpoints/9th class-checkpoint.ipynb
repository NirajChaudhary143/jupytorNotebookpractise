```python
class college:
    def __init__ (self,college_name,college_address):
        self.college_name=college_name
        self.college_address=college_address
    def CollegeName(self):
        print(self.college_name)
    def CollegeAddress(self):
        print(self.college_address)
clg = college("Nagarjuna College","Sankhamul")
clg.CollegeName()
clg.CollegeAddress()
```
